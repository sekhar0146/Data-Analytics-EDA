```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# sklearn
import seaborn as sns
```

```python
# Import the data
df = pd.read_csv("C:/Users/z011348/Desktop/ML/input/titanic/train.csv")
df
```

```python
# which age passingers are travelled (histogram)
plt.figure(figsize=(10,6))
df.Age.hist();
```

```python
# how many people survived (0 = No, 1 = Yes) - Count
sns.set_style("darkgrid")
#1. diaplay the survived info 
plt.figure(figsize=(7,5))
sns.countplot(x='Survived', data=df, palette='Set2')
#2. Rotate x-labels
plt.xticks(rotation=0);
```

```python
# Survival report on genderwise
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.countplot(x="Survived", hue='Sex', data=df, palette='spring');
```

```python
# Survial report on classwise
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.countplot(x="Survived", hue='Pclass', data=df, palette='muted');
```

```python
# Survial report on age
plt.figure(figsize=(13,8))
sns.set_style("darkgrid")
sns.countplot(x="Survived", hue='Age', data=df, palette='Paired');
```

```python
# Survial report on Siblings
#sibsp: The dataset defines family relations in this way...
#Sibling = brother, sister, stepbrother, stepsister
#Spouse = husband, wife (mistresses and fiancÃ©s were ignored)

plt.figure(figsize=(13,8))
sns.set_style("darkgrid")

sns.countplot(x="Survived", hue='SibSp', data=df, palette='deep');
```

```python
# Survial report on Parent-Child relationship
#parch: The dataset defines family relations in this way...
#Parent = mother, father
#Child = daughter, son, stepdaughter, stepson
#Some children travelled only with a nanny, therefore parch=0 for them.

plt.figure(figsize=(13,8))
sns.set_style("darkgrid")
sns.countplot(x="Survived", hue='Parch', data=df, palette='Paired');
```

```python
# Find out the average age of survived or not using boxplot
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.boxplot(x="Survived", y="Age", palette=["m", "g"], data=df)
sns.despine(offset=10, trim=True)
```

```python
# Find out the average age of survived or not per each class using boxplot
plt.figure(figsize=(13,9))
sns.set_style("darkgrid")
sns.boxplot(x="Survived", y="Age", hue='Pclass', data=df)
sns.despine(offset=10, trim=True)
```
